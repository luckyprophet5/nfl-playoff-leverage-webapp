```python
page = requests.get(url)
```

```python
soup = BeautifulSoup(page.text,'html.parser')
```

```python
events = soup.find_all('tbody', class_='sportsbook-table__body')
games = []
fucked_events = []
for event in events:
    away = True
    game_dict = {}
    for outcome_cell in event.find_all('div',class_='sportsbook-outcome-cell'):
        cell_body = outcome_cell.find(class_='sportsbook-outcome-cell__body')
        outcome = outcome_cell.find(class_='sportsbook-outcome-cell__line')
        odds = outcome_cell.find(class_='sportsbook-odds')
        split = cell_body['aria-label'].split()
        if len(split) == 3:
            team = f"{split[0]} {split[1]}"
            try:
                print(f"{team} {outcome.text} {odds.text}")
                if away:
                    game_dict['away_team'] = team
                    game_dict['away_spread'] = float(outcome.text)
                    game_dict['away_spread_odds'] = float(odds.text)
                    away = False
                else:
                    game_dict['home_team'] = team
                    game_dict['home_spread'] = float(outcome.text)
                    game_dict['home_spread_odds'] = float(odds.text)
            except AttributeError:
                fucked_events.append(outcome_cell)
                if away:
                    away = False
    games.append(game_dict)
```
